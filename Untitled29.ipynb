```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.ensemble import AdaBoostClassifier

# Load the dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
df = pd.read_csv(url, names=names)

# Split the dataset into features and target variable
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Split the dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Standardize the features
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Train the AdaBoost classifier
clf = AdaBoostClassifier(n_estimators=100, learning_rate=1, random_state=1)
clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python

import matplotlib.pyplot as plt

# Calculate the accuracy for each boosting round
accuracy = []
for i in range(1, 101):
    clf = AdaBoostClassifier(n_estimators=i, learning_rate=1, random_state=1)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    accuracy.append(accuracy_score(y_test, y_pred))

# Plot the accuracy for each boosting round
plt.figure(figsize=(12, 6))
plt.plot(range(1, 101), accuracy, marker='o', linestyle='--', color='blue')
plt.title('AdaBoost Classifier')
plt.xlabel('Boosting Rounds')
plt.ylabel('Accuracy')
plt.show()
```
